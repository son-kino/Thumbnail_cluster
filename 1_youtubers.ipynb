```python
def channel_split(data):
    channel_data = {}
    for line in data.split('\n'):
        parts = line.split()
        if len(parts) < 5:
            # 데이터 구성이 예상과 다르므로 건너뜁니다.
            continue
        rank = parts[0]
        rank_change = parts[1] if parts[1].isdigit() else None
        category = parts[-5]
        name = ' '.join(parts[2:-5])

        subscribers = parts[-4]
        subscriber_change = parts[-3]
        avg_views = parts[-2]
        avg_views_change = parts[-1]

        channel_data[name] = {
            "순위": rank,
            "순위_변화": rank_change,
            "이름": name,
            "카테고리": category,
            "구독자_수": subscribers,
            "변화율": subscriber_change,
            "평균_조회수": avg_views,
            "조회수_변화율": avg_views_change
        }

    return channel_data
```

```python
import csv
def save_to_csv(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ["순위", "순위_변화", "이름", "카테고리", "구독자_수", "변화율", "평균_조회수", "조회수_변화율"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        
        writer.writeheader()
        for channel in data.values():
            writer.writerow(channel)
```
